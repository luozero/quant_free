### Graphical Models

$$\text{OI}_t := f_1 \left[p_t - \nu_t\right] + \epsilon_{1,t}$$

$$p_{t+h} - \nu_t := f_2\left[ \text{OI}_t\right] + f_3 \left[ \text{MOM}_t\right] + \epsilon_{2,t+h}$$
$$\text{HML}_t := f_4 \left[ \text{MOM}_t\right] + \epsilon_{3,t}$$

```python
import graphviz as gr

dummy_model = gr.Digraph(
    'G', filename = 'dummy_model.gv',
    node_attr = {'shape': 'plaintext'}, 
) 

with dummy_model.subgraph(name = 'cluster_0') as c:
    c.attr(style = 'filled', color = 'lightgray')
    c.node('HML')
    c.node('Order Imbalance')
    c.node('Price Convergence')
    c.attr(rank = 'same')

dummy_model.edge('Momentum', 'HML')
dummy_model.edge('Momentum', 'Price Convergence')

dummy_model.edge('HML', 'Order Imbalance')
dummy_model.edge('Order Imbalance', 'Price Convergence')

dummy_model.edge('Order Imbalance', 'Bid Ask Spread')
dummy_model.edge('Price Convergence', 'Bid Ask Spread')

dummy_model.edge('Bid Ask Spread', 'Liquidity')
dummy_model.edge('Bid Ask Spread', 'Volatility')
```
