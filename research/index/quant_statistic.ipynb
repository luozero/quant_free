```python
# Set font to DejaVu Sans or SimHei
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']  # or 'SimHei' for Chinese font support
plt.rcParams['axes.unicode_minus'] = False

# Proceed with your seaborn/matplotlib plotting
sns.set()

df= us_dir1_load_csv(dir0 = 'symbol', dir1 = 'xq', filename='xq_index_price_ratio.csv')
df['date'] = pd.to_datetime(df['date'])

# Keep only the date part
df['date'] = df['date'].dt.date
df['date'] = pd.to_datetime(df["date"], errors='coerce')
df.set_index('date', inplace=True)
ret = df

metrics = qs.reports.metrics(ret, mode='basic', display = False, as_pct = True)

metrics = metrics.T.sort_values(by='CAGR% (Annual Return)', axis=0, ascending=False).T

us_dir1_store_csv(dir0 = 'symbol', dir1 = 'xq', filename='xq_index_analysis' + '.csv', data = metrics, encoding='utf-8-sig')

display(metrics)
# Display DataFrame in a scrollable box
# output = widgets.Output()
# with output:
#     display(metrics)
    
# scrollable = widgets.VBox([output])
# display(scrollable)
```
